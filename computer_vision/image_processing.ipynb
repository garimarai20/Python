FILTERS

```python
filters = [
   ImageFilter.CONTOUR,
   ImageFilter.BLUR,
   ImageFilter.DETAIL,
   ImageFilter.EDGE_ENHANCE,
   ImageFilter.EDGE_ENHANCE_MORE,
   ImageFilter.EMBOSS,
   ImageFilter.FIND_EDGES,
   ImageFilter.SHARPEN,
   ImageFilter.SMOOTH,
   ImageFilter.SMOOTH_MORE
]

for i,f in enumerate(filters):
    im2.filter(f).save(f'generated/filter_{i}.png')
```

```python
im2.filter(ImageFilter.MaxFilter(11)).show()
im2.filter(ImageFilter.MinFilter(11)).show()
im2.filter(ImageFilter.ModeFilter(11)).show()
im2.filter(ImageFilter.MedianFilter(11)).show()
```
