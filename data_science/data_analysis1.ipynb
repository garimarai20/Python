- Selecting rows and columns 
    - ROWS 
       - 'loc'  - gets us the data using named indexes
       - 'iloc' - gets us the data using numeric indexes
    - COLUMNS 
       - 'df[single]' - requires name of column 
       - 'df[multiple]' - requires list of column ( multiple is a list of named columns )

```python
# Single row ( record )
weather.iloc[0]  # first record 
```

```python
# Single row ( record )
weather.iloc[-1] # last record 
```

```python
# Multiple records ( 25 - 50)
weather.iloc[25:50]
```

```python
# records 1- 15  with first 5 cols
weather.iloc[1:16,[0,1,2,3,4]]
```

```python
weather.iloc[:15,-3:]
```

```python
weather['Date']  # selecting a single column ( dictionary style )  --> Series
```

```python
weather.AvgTemp  # same as weather['AvgTemp']  ( Object style)
```

```python
cols = [ 'Date', 'MaxTemp', 'MinTemp']
weather[cols]  # multiple cols ( dictionary style)  --> DataFrame
```

```python
weather[ ['Precipitation', 'State' , 'City' , 'Date']] # shorter version
```
