```python
filepath ='Placement_Data_Full_Class.csv'
if not os.path.exists(filepath):
    print('File not found')
else:
    df= pd.read_csv(filepath, index_col= 0)
    print(df.info())
```

```python
df['hsc_s'].value_counts().plot(kind='bar',title='High school specialisation distribution',figsize=(15,6))
```

```python
df.plot(kind='hexbin', x='salary', y='etest_p',colormap='plasma')
```

```python
df.plot(kind='hist', y='salary',colormap='plasma')
```

```python
df['gender'].value_counts().plot(kind='pie',title='gender distribution',figsize=(15,6), colormap='rainbow')
```

```python
df['status'].value_counts().plot(kind='bar',title='PLACED ',figsize=(15,6),colormap='Pastel2')
```

```python
df.pivot_table(values='salary',index='degree_t',columns='gender')
```

```python
df.pivot_table(values='salary',index='degree_t',columns='gender').plot(kind='bar',colormap='copper_r')
```
